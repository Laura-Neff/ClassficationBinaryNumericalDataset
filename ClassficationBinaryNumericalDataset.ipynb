```python
data = pd.read_csv('gender_voice_dataset.csv')

data.sample(10)
```

##Label encoding

```python
from sklearn.preprocessing import LabelEncoder

labelEncoder = LabelEncoder()
data['label'] = labelEncoder.fit_transform(data['label'].astype(str))

data.head()
```

```python
data.boxplot(by ='label', column =['meanfreq'], grid = False, figsize=(10, 8))
```

##Spilting the data into train and test data

```python
from sklearn.model_selection import train_test_split

X = data.drop('label', axis=1)
Y = data['label']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0) 
```

```python
x_train.shape , y_train.shape
```

```python
x_test.shape , y_test.shape 
```

##Logistic Regression Classifier
https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html



```python
from sklearn.linear_model import LogisticRegression

logistic_model = LogisticRegression(penalty='l2', solver='liblinear')
logistic_model.fit(x_train, y_train)

#simplest, most powerful classification technique is logistic regression
#penalty is l2 so we have a little bit of regularization when we fit the model
#the solver is the optimization algorithm that logistic regression uses under the hood 
#liblinear is usually good for small datasets like ours
```

```python
y_pred = logistic_model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix = confusion_matrix(y_test, y_pred)

print(confusion_matrix)

#usually useful and easy to visualize classification results on a confusion matrix 

#257 and 323, the main diagonal numbers, represent the number of predictions in the test data that our model got right
```

```python
#the most common score to evaluate a classification model is the accuracy of the model
print("Training score : ", logistic_model.score(x_train, y_train))
```

Accuracy, Precision , recall Score

```python


from sklearn.metrics import accuracy_score, precision_score, recall_score

acc = accuracy_score(y_test, y_pred)
pre = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)

print('Accuracy : ' , acc) #the number of voices correctly classified 
print('Precision Score : ', pre) #of all voices classified as male i.e positive identifications, how many were classified correctly?
print('Recall Score : ', recall) #of all male voices in the dataset, how many were classified correctly as male?
```
